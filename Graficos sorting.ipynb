```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
heapSort = [0.0004892349243164062, 0.0024793148040771484, 0.030553579330444336]
quickSort = [0.0026009082794189453, 0.0795295238494873, 26.569446325302124]
```

```python
barWidth = 0.25

plt.figure(figsize=(10,5))

r1 = np.arange(len(heapSort))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
```

```python
plt.bar(r1,heapSort, color='#6A5ACD',width=barWidth, label='HeapSort')
plt.bar(r2,quickSort, color='#6495ED',width=barWidth, label='QuickSort')
#plt.bar(r1,heapSort, color='#6A5ACD',width=barWidth, label='HeapSort')

plt.xlabel('Quantidade de elementos')
plt.xticks([r + barWidth for r in range(len(heapSort))], ['100 elementos, 1000 elementos, 10000 elementos, 100000 elementos'])
plt.ylabel('Segundos')
plt.title('Comparação de algoritmos de ordenação pela quantidade de elementos (Quesito A)')

plt.legend()
plt.show
```
